```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql+psycopg2://postgres:root@localhost:5432/BankingTransactionAnalysis(NovaBank)")

branches = pd.read_sql("select * from cleaned_branches", engine)
customers = pd.read_sql("select * from cleaned_customers", engine)
accounts = pd.read_sql("select * from cleaned_accounts", engine)
transactions = pd.read_sql("select * from cleaned_transactions", engine)

print(f"branches loaded with {len(branches)} rows")
print(f"customers loaded with {len(customers)} rows")
print(f"accounts loaded with {len(accounts)} rows")
print(f"transactions loaded with {len(transactions)} rows")
```

```python
# Core KPI — Customer Retention Rate
```

```python
total_customers = len(customers)
new_customers = len(customers[customers['lifecycle_stage'] == 'New'])
active_customers = len(customers[customers['lifecycle_stage'] == 'Active'])

# Exclude new customers from retention denominator
retention_base = total_customers - new_customers
retention_rate = round((active_customers / retention_base) * 100, 2)

print(f"Total Customers:     {total_customers:,}")
print(f"New Customers:       {new_customers:,}")
print(f"Active Customers:    {active_customers:,}")
print(f"Retention Base:      {retention_base:,}")
print(f"Retention Rate:      {retention_rate}%")
```

```python
active_customers = len(customers[customers['lifecycle_stage'] == 'Active'])
dormant_customers = len(customers[(customers['lifecycle_stage'] == 'Dormant')&(customers['is_non_transacting_customer']==False)]) # Here we will get actual customers who hasn't transacted scince 6 months
churned_customers = len(customers[
    (customers['is_non_transacting_customer'] == False) &
    (customers['lifecycle_stage'] == 'Dormant') &
    (customers['days_since_last_trans'] > 365)
])
non_transacting_customers = len(customers[customers['is_non_transacting_customer'] == True]) # New Metric

print(f"Active Customers:            {active_customers:,}")
print(f"Dormant Customers:           {dormant_customers:,}")
print(f"Churned Customers:           {churned_customers:,}")
print(f"Non Transacting Customers:   {non_transacting_customers:,}")
```

```python
# 1. Map existing rfm_segment categories to the three requested target groups
mapping = {
    'High Value': 'High-value',
    'Medium Value': 'Medium-value',
    'Low Value': 'Low-engagement',
    'At Risk': 'Low-engagement',
    'Dormant': 'Low-engagement'
}

# 2. Map the categories to a new column
customers['target_segment'] = customers['rfm_segment'].map(mapping)

# 3. Calculate absolute counts and percentages
counts = customers['target_segment'].value_counts()
percentages = customers['target_segment'].value_counts(normalize=True) * 100

# 4. Display the results in a summary table
summary_df = pd.DataFrame({
    'Customer Count': counts,
    'Percentage (%)': percentages.round(2).astype(str) + '%'
})

print(summary_df)
```

```python
# Cohort Analysis 
```

```python
cohort_df = transactions[transactions['transaction_status']=='Successful'].merge(
    customers[['customer_id','signup_date']], 
    on='customer_id', 
    how='left'
)
```

```python
cohort_df['transaction_datetime'] = pd.to_datetime(cohort_df['transaction_datetime'])
cohort_df['signup_date'] = pd.to_datetime(cohort_df['signup_date'])
cohort_df['cohort'] = cohort_df['signup_date'].dt.to_period('M')
cohort_df['months_since_signup'] = (
    (cohort_df['transaction_datetime'].dt.to_period('M')) - (cohort_df['signup_date'].dt.to_period('M'))
).apply(lambda x: x.n)
```

```python
cohort_counts = cohort_df.groupby(['cohort', 'months_since_signup']
)['customer_id'].nunique().reset_index()

# Step 4b — pivot into matrix
cohort_matrix = cohort_counts.pivot_table(
    index='cohort',
    columns='months_since_signup',
    values='customer_id'
)

# Step 4c — convert to retention percentage
# Hint: divide every row by its Month 0 value
cohort_size = cohort_matrix[0]  # Month 0 = cohort size
retention_matrix = cohort_matrix.divide(cohort_size, axis=0) * 100

print(retention_matrix.round(1))
```

```python
# 1. Correct cohort size — all customers who signed up that month
customer_cohorts = customers.copy()
customer_cohorts['signup_date'] = pd.to_datetime(customer_cohorts['signup_date'])
customer_cohorts['cohort'] = customer_cohorts['signup_date'].dt.to_period('M')

# 2. Cohort sizes — total signups per month
cohort_sizes = customer_cohorts.groupby('cohort')['customer_id'].nunique()

# 3. Rebuild cohort counts from transactions
cohort_counts = cohort_df.groupby(
    ['cohort','months_since_signup']
)['customer_id'].nunique().reset_index()

cohort_matrix = cohort_counts.pivot_table(
    index='cohort',
    columns='months_since_signup',
    values='customer_id'
)

# ---- FIXED ORDER STARTS HERE ----

# 4. Calculate retention matrix first
retention_matrix = cohort_matrix.divide(cohort_sizes, axis=0) * 100
retention_matrix = retention_matrix.round(1)

# 5. Filter for cohorts up to 2023-11 now that the matrix exists
retention_matrix = retention_matrix[
    retention_matrix.index <= '2023-11'
]

# 6. Print the first 12 months of the cleaned matrix
print("--- Cohort Retention Matrix (%) ---")
print(retention_matrix.iloc[:, :12])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 10))
sns.heatmap(
    retention_matrix.iloc[:, :12],
    annot=True,
    fmt='.0f',
    cmap='RdYlGn',
    vmin=0,
    vmax=100,
    linewidths=0.5,
    cbar_kws={'label': 'Retention %'}
)
plt.title('NovaBank — Customer Cohort Retention Matrix (%)', 
          fontsize=14, pad=15)
plt.xlabel('Months Since Signup')
plt.ylabel('Signup Cohort')
plt.tight_layout()
plt.savefig('cohort_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```
